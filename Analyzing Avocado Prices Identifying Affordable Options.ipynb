### Clearly the above boxplot indicates that all the variables contains outliers. Now we need to take care of the outliers.

### <font color='orange'>Before we go on to taking care of the outliers we removed the "Date" variable from our dataset as it is useless now.</font>

```python
avocado_df.drop(columns=["Date"],inplace=True)
avocado_df.info()
```

### <font color='orange'>In the below code snippet, we will replace the outliers higher than the upper whisker by the value of the upper whisker and the outliers lower than the lower whisker by the value of the lower whisker.</font>

```python
# Replace the outliers
columns = avocado_df.columns
for j in columns:
    if isinstance(avocado_df[j][0], str) :
        continue
    else:
        #defining quartiles
        quartiles = percentile(avocado_df[j], [25,75])
        # calculate min/max
        lower_fence = quartiles[0] - (1.5*(quartiles[1]-quartiles[0]))
        upper_fence = quartiles[1] + (1.5*(quartiles[1]-quartiles[0]))
        avocado_df[j] = avocado_df[j].apply(lambda x: upper_fence if x > upper_fence else (lower_fence if x < lower_fence else x))
```

### <font color='orange'>Define function that computes various metrics for regression models</font>

```python
def regression_results(y_true, y_pred):
    # Calculate the explained variance score
    explained_variance=metrics.explained_variance_score(y_true, y_pred)
    # Calculate the mean absolute error
    mean_absolute_error=metrics.mean_absolute_error(y_true, y_pred)
    # Calculate the mean squared error
    mse=metrics.mean_squared_error(y_true, y_pred) 
    # Calculate the R^2 score
    r2=metrics.r2_score(y_true, y_pred)
    # Calculate the adjusted R^2 score
    adjusted_r2 = 1 - (1-r2)*(len(y_true)-1)/(len(y_true)-X_test.shape[1]-1)

    # Print the results
    print('Explained_variance: ', round(explained_variance,4))    
    print('R2: ', round(r2,4))
    print('Adjusted_r2: ', round(adjusted_r2,4))
    print('MAE: ', round(mean_absolute_error,4))
    print('MSE: ', round(mse,4))
    print('RMSE: ', round(np.sqrt(mse),4))
```

### <font color='orange'>Define function that uses cross-validation to estimate the accuracy of a given machine learning model.</font>

```python
def model_accuracy(model,X_train=X_train,y_train=y_train):
    # Using cross-validation to evaluate the accuracy of the mode
    accuracies = cross_val_score(estimator = model, X = X_train, y = y_train, cv = 10)
    # printing out the mean accuracy
    print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
    # printing out the standard deviation of the accuracy 
    print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```
